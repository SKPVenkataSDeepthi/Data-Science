Time series analysis and forecasting are crucial for predicting future trends, behaviors, and behaviours based on historical data.

## What is time series? 
* sequence of data points collected, recorded at successive and evenly spaced time intervals. 
* each data point represents observations take at that time.
* mostly reresented by graph (as a line dot) with time on horizontal axis and its varibale on vartical axis.


## Advantages of time series analysis in business: 
* Mitigate risks 
* informed demands
* optimized resources

## Importance of time series analysis
* predicting future trends.
* Detect patterns and anamalies
* Risk mitigation (spotting potential risks)
* strategic planning (guiding decision making across various sectors).
* optimize resource allocation effectively (competitive edge)

## Components of time series analysis 
1. Trend: rep. overall movement or directionality of the data over time. These can be linear(showing consistent increase or decrease) and nonlinear(showing more complex patterns). 
2. Seasonality: refers to periodic fluctuations or patterns that occur at regular intervals. These repeat annually , quarterly, monthly or weekly. 
3. Cyclicality (Cyclic variations): These are the longer term fluctuations in time series, these do not have fixed term 
4. Irregular (Noise): It is also known as  noice or randomness, refers to unpredictable random fluctuations in data, these fluctuations may result from random events, measurement errors or from other foreseen factors. Irregualrity makes it challenging to identify and model underlying patterns in time series data.

## Visualization of time series
These visualization methods helps to identify trends, patterns and anamalies in time dependent data. 
1. Line PLots: display data points over time, easy observation of trends, cycles and fluctuations. 
2. Seasonal plots: these break time series data into seasonal components, helps in visualizing patterns within specific time period. 
3. Histograms & density plots: dist. of data values over time, provides insights on skewness and kurtosis (kurtosis is that describes the shape of a distribution's tails in relation to its overall shape, it's often used to assess the presence of outliers or the extremity of data points in a dataset). 
4. Autocorrelation and partial correlation: visualize correlation between time series and lagged values. helps in identifying lagged relationships and seasonality. 
5. Spectral analysis: visualize frequency components within time series data, useful in identifying periodicity and cyclical patterns. 
6. Decomposition plots: these break down a time series into its trend, seasonal and residual components

## Seasonal Detection in Time Series Data

It plays a cruicial role revealing recurring patterns within data. 

### What is seasonality? 
It refers to predictable patterns that occur at regualar intervals within a time series. 
* Identifying seasonality is crucial for understanding inherent structure of data and can aid in informed decisions.

### Why should we detect seasonality in time series data? 
* To understand recurring patterns( pattern recognition).
* Components significantly impact forecasting accuracy (forecasting).
* Identifying irregularities in data (Anamaly detection).
* Optimized decision making. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose ### The seasonal_decompose function from the statsmodels.tsa.seasonal module is used for decomposing time series data into its three main components: trend, seasonality, and residuals.

# Load the dataset
data = pd.read_csv('AirPassengers.csv')
data['Month'] = pd.to_datetime(data['Month'], format='%Y-%m')
data.set_index('Month', inplace=True)
 
# Plot the original time series data
plt.figure(figsize=(7, 5))
plt.plot(data, label='Original Time Series')
plt.title('Air Passengers Time Series')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
### Data decomposition

# Decompose the time series into trend, seasonal and residual components
result = seasonal_decompose(data, model='multiplicative', extrapolate_trend='freq')
### Data: rep. the time series data that we want to decompose. 
### model: specifies the type of decomposition to be performed this takes 2 values addictive and multiplicative. Multiplicative is used to see the amplitude of seasonal component(pattern is constant), here seasonal and trend components are multiplied rather than added (additive model..)
### extrapolate_model: this controls whether to extrapolate trend component to cover missing values at end of time series. Here it is set to freq means that trend component is extrapolated using the frequency of time series. This can be used when there are missing values at the end of time series. 
result.plot()
plt.suptitle('Seasonal Decomposition of Air Passengers Time Series')
plt.tight_layout()
plt.show()
```

```python
## visualizing the seasonality 
# Plot the seasonal component
plt.figure(figsize=(6, 4))
plt.plot(result.seasonal, label='Seasonal Component')
plt.title('Seasonal Component of Air Passengers Time Series')
plt.xlabel('Year')
plt.ylabel('Seasonal Component')
plt.legend()
plt.show()
```

```python
#### Removing seasonality from the data

### this is the important step to do before proceeding to model training phase because, seasonality can degrade the performance of predictive model which leads to wrong forecasting. 

# Plotting the original data and original data without the seasonal component
plt.figure(figsize=(7, 4))
# Plot the original time series data
plt.plot(data, label='Original Time Series', color='blue')
data_without_seasonal = data['#Passengers'] / result.seasonal
# Plot the original data without the seasonal component
plt.plot(data_without_seasonal, label='Original Data without Seasonal Component', color='green')
plt.title('Air Passengers Time Series with and without Seasonal Component')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

## Time Series Decomposition Techniques

It is a process of seperating atime series into its consequent components, such as trend, seasonality, and noise. 
* this helps us breakdown a time series dataset into three main components.
* 1. Trend: long term movement in the data
  2. Seasonality: short term fluctuations caused by factors like seasons and cycles.
  3. Residual(Noise): rep. random variablity that remains after removing trend and seasonality.
* By seperating these components we can gain insights into behaviour of the data and make better forecasts.
  
## Types of Time Series Decomposition Techniques: 
1. Additive decomposition: time series expressed as sum of its components
   Y(t) = Trend(t) + Seasonalt(t) + Residual(t)
   suitable when magnitude of seasonality doesnt vary with magnitude of time series. 
3. Multiplicative decomposition: time series expressed as product of its components.
   Y(t) = Trend(t) * Seasonalt(t) * Residual(t)
   suitable when magnitude of seasonality scales with magnitude of time series.

## Methods of Decomposition. 
1. Moving Averages: cal. avg. of certain number of past data points.
2. Seasonal Decomposition of Time series: STL - Seasonal and Trend Decomposition using Loess, it is a popular method for decomposition. It uses local regression (Loess) to extract the trend and seasonality components.
3. Exponential Smoothing State Space Model: it uses ETS(Error, Trend, Seasonal) framework to estimate trend and seasonal components. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

## creating synthetic time series dataset
np.random.seed(0)
date_rng = pd.date_range(start="2021-01-01", periods=365, freq="D")
data = np.sin(np.arange(365) * 2 * np.pi / 365) + np.random.normal(0, 0.5, 365)
ts = pd.Series(data, index=date_rng)
ts
```

```python
## visulaizing the dataset
plt.figure(figsize=(12, 3))
plt.plot(ts, label='Original Time Series')
plt.legend()
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Create a pandas Series
x = pd.Series(data, index=date_rng)

# Decompose the time series using an additive model
result_add = seasonal_decompose(x, model='additive')

# Plot the additive trend component
plt.figure(figsize=(10, 6))
plt.plot(result_add.trend, label='Additive Trend', color='blue')
plt.legend()
plt.title('Additive Trend Component')
plt.xlabel('Date')
plt.ylabel('Trend')
plt.show()
```

```python
### plotting seasonal component
plt.figure(figsize=(9, 3))

plt.plot(result_add.seasonal, label='Additive Seasonal')
plt.legend()
```

```python
## calculating Simple Moving Average (SMA). 
sma_window = 7  # 7-day moving average
sma = ts.rolling(window=sma_window).mean()
sma
```

```python
## Calculating Exponential Moving Average
ema_window = 30  # 30-day moving average
ema = ts.ewm(span=ema_window, adjust=False).mean()
ema
```

```python
## plotting the both moving averages
plt.figure(figsize=(9, 3))
plt.plot(ts, label='Original Time Series')
plt.plot(sma, label=f'{sma_window}-Day SMA')
plt.plot(ema, label=f'{ema_window}-Day EMA')
plt.legend()
```

Moving averages are used for smoothing out data and simplifying complex patterns.
