Unsupervised Learning deals with unlabeled data. 

## TASKS: 
* Finding patterns
* Finding relationships

> Powerful tool for EDA Exploratory Data Analysis
> IN AI, ML that takes place in absence of human supervision is known as unsupervised machine learning.
> The training model has input parameters values and discovers the groups or patterns on its own.

## OUTLINE
Input Raw Data ---> Model Training ---> Output 
* Model Training
Interpretation(unknown output, no training dataset)---> Algorithm ----> Processing.

## WORKING: 
The data is not labelled with any preferred categories or outcomes, so the algorithm must find these patterns and relationships on its own

## INPUTS OF THE DATA
1. Unstructured data: may contain nosiy data, missing values, unknown data. 
2. Unlabeled data: contains a value for input parameters, no target value

## Unsupervised Algorithms
1. Clustering
2. Association Rule Learning
3. Dimensionality Reduction.

## 1. Clustering
Process of grouping unlabelled data into clusters based on similarities. 
GOAL: Identify patterns and relationships to the data without prior knowledge of datas meaning. 
* This is used to group data based on different patterns.
* These algorithms are used to process raw, unclassified dataand objects into groups.
## Common Clustering Algorithms.
1. K-Means Clustering
2. Hierarchial Clustering
3. Density-Based Clustering
4. Mean-Shift Clustering
5. Spectral Clustering

## 2. Association Rule Learning. 
* Also known as Association Rule Mining
* Common technique used to find relations between parameters of large dataset.
## Common Association Rule Learning Algorithms
1. Apriori Algorithm: classic method for rule induction
2. FP-Growth Algorithm: efficient alternative to apriori 
3. Eclat Algorithm: exploiting closed items for fficient rule mining
4. Efficient Tree-based Algorithm: handling large datasets with scalability.

## 3. Dimensionallity Reduction
* Process of reducing the number of features in dataset while preserving as much information as possible.
* Imporves performance of ML algorithm, data visualization.
## Types
1. PCA - Principle Component Analysis: Linear Transformation for reduced dimensions
2. LDA - Linear Discriminant Analysis: dim. red. for discrimination.
3. NMF - Non-Negative Matrix Factorization: Decomposing data into non negative components
4. LLE - Locally Linear Embedding - Preserving Local Geometry in reduced dimensions
5. Isomap: Capturing global relationships in reduced dimensions.

## Challenges of Unsupervised Learning: 
1. Evaluation: without labelles or categories accessing performance is difficult
2. Interpretability: understanding decison making process is challenging
3. Overfitting: they can overfit to some datasets lin=miting their abaility to generalize to new data. 
4. Data Quality: sensitive to quality of input data. 
5. Computational Complexity: those dealing with high dimensional data or large datasets can be computationally expensive.

## Advantages of Unsupervised Learning: 
1. Can uncover hidden patterns
2. can be used to explaore new data

## Disadvantages of Unsupervised Learning: 
1. Difficult to evalute
2. sensitive to quality of data
3. computationally expensive
4. difficult to interpret
